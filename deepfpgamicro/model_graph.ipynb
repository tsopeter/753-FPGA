```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('filtered_models.csv')
valid_data = df[df['Valid'] == True]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Weights', y='BRAM', data=valid_data, s=100)

plt.xlabel('Weights')
plt.ylabel('BRAM Usage')
plt.title('BRAM Usage vs Weights for Valid Configurations')

plt.grid(True, linestyle='--', alpha=0.7)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Weights', y='LUT', data=valid_data, s=100)

plt.xlabel('Weights')
plt.ylabel('LUTs')
plt.title('LUTs vs Weights for Valid Configurations')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Connections', y='BRAM', data=valid_data, s=100)

plt.xlabel('Connections')
plt.ylabel('BRAM')
plt.title('BRAM usage vs Connections for Valid Configurations')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Connections', y='LUT', data=valid_data, s=100)

plt.xlabel('Connections')
plt.ylabel('LUTs')
plt.title('LUTs vs Connections for Valid Configurations')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
