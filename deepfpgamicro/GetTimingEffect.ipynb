```python
from util import networkValues
from matplotlib import pyplot as plt
```

```python
periods = [10, 20, 30, 50, 70, 100]
reports = []
for i, period in enumerate(periods):
    mr = networkValues("1101", "000", 0.1, 6, period, limits=1.0)
    reports.append(mr.report)

```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(reports)
# Ensure TargetClock is sorted numerically
df_sorted = df.sort_values("TargetClock")

x = np.arange(len(df_sorted))
width = 0.2
labels = df_sorted["TargetClock"].astype(str)

fig, ax1 = plt.subplots(figsize=(12, 6))

# Primary y-axis (LUT and FF)
bar1 = ax1.bar(x - width, df_sorted["LUT"], width, label="LUT")
bar2 = ax1.bar(x, df_sorted["FF"], width, label="FF")
ax1.set_ylabel("LUT / FF")
ax1.set_xlabel("Target Clock (ns)")
ax1.set_xticks(x)
ax1.set_xticklabels(labels, rotation=45)
ax1.legend(loc="upper left")
ax1.grid(True, axis='y')

# Secondary y-axis (BRAM and DSP)
ax2 = ax1.twinx()
bar3 = ax2.bar(x + width, df_sorted["BRAM_18K"], width, label="BRAM_18K", color="orange")
bar4 = ax2.bar(x + 2*width, df_sorted["DSP"], width, label="DSP", color="green")
ax2.set_ylabel("BRAM_18K / DSP")
ax2.legend(loc="upper right")

plt.title("Resource Usage by Target Clock")
plt.tight_layout()
plt.show()
```
